```python
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
root_path=os.getcwd()
mask_path=root_path+'/test_imgs/mask.png'
foreground_path=root_path+'/test_imgs/foreground.jpg'
backround_path=root_path+'/test_imgs/background.jpg'
mask_image =Image.open(mask_path)
foreground_image =Image.open(foreground_path)
background_image =Image.open(backround_path)
```

```python
mask_pixel=mask_image.load()
foreground_pixel=foreground_image.load()
background_pixel=background_image.load()
```

```python
count=0
for i in range(mask_image.size[0]):
    for j in range(mask_image.size[1]):
        if mask_pixel[i,j]==[255,0,0]:
            background_pixel[i,j]=foreground_pixel[i,j]
        count=count+1
```
